```python
import os
from sklearn.pipeline import Pipeline
from transform import get_pipeline_steps
import pandas as pd
```

```python
years = list(range(2008, 2020))
skip_rows = [8] * len(years)
skip_rows[-1] = 7

all_df = []
for year, skip_row in zip(years, skip_rows):
    data_path = os.path.join('data', f'{year}', 'Forma07.xls')
    steps = get_pipeline_steps(skip_row, year)
    pipeline = Pipeline(steps=steps)
    df = pipeline.transform(data_path)
    all_df.append(df)
    print(f'year {year} ok')
```

```python
all_df[0].head()
```

```python
df = pd.concat(all_df)
```

```python
df.to_excel(os.path.join('out', 'cancer-ua-2008-2019.xls'), index=False)
```
